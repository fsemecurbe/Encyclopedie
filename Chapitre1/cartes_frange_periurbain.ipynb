```python
import pandas as pd
import geopandas as gpd
import numpy as np
import contextily as ctx
import matplotlib.pyplot as plt
from matplotlib_scalebar.scalebar import ScaleBar
import shutil
```

```python
shutil.unpack_archive('peribesancon.zip')
shutil.unpack_archive('perilille.zip')
```

```python
besancon = gpd.read_file('peribesancon.shp')
lille = gpd.read_file('perilille.shp')
```

```python
x, y=  695997.789,7055437.341
ax = lille.plot( color='Black', markersize=0.1, figsize=(10,10))
#ctx.add_basemap(ax,   crs=lille.crs.to_string(), source=ctx.providers.OpenStreetMap.Mapnik, zoom=14)
ax.set_axis_off()
ax.set_xlim((x - 2500, x + 2500))
ax.set_ylim((y - 2500, y + 2500))
ax.add_artist(ScaleBar(1, location='lower right'))
plt.savefig('periurbain_lille.png')
```
