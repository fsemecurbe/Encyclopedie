```python
import pandas as pd
import geopandas as gpd
import numpy as np
import contextily as ctx
import matplotlib.pyplot as plt
from matplotlib_scalebar.scalebar import ScaleBar
```

```python
bocage = pd.read_csv('bocage_ign_2012.csv')
bocage = gpd.GeoDataFrame(bocage, geometry=gpd.points_from_xy(bocage.x, bocage.y), crs="EPSG:2154")
openfield = pd.read_csv('openfield_ign_2012.csv')
openfield = gpd.GeoDataFrame(openfield, geometry=gpd.points_from_xy(openfield.x, openfield.y), crs="EPSG:2154")
```

```python
ax = bocage.plot( color='Black', markersize=0.1, figsize=(10,10))
ctx.add_basemap(ax,   crs=bocage.crs.to_string(), source=ctx.providers.OpenStreetMap.Mapnik)
ax.set_axis_off()
ax.set_xlim((bocage.x.min(), bocage.x.max()))
ax.set_ylim((bocage.y.min(), bocage.y.max()))
ax.add_artist(ScaleBar(1, location='lower right'))
plt.savefig('bocage_large.png')
```

```python
ax = bocage.plot( color='Black', markersize=1, figsize=(10,10))
ctx.add_basemap(ax,zoom=13, crs=bocage.crs.to_string(), source=ctx.providers.OpenStreetMap.Mapnik)
ax.set_axis_off()
ax.set_xlim((388000, 398000))
ax.set_ylim((6890000, 6900000))
ax.add_artist(ScaleBar(1, location='lower right'))
plt.savefig('bocage_small.png')
```

```python
ax = openfield.plot( color='Black', markersize=0.1, figsize=(10,10))
ctx.add_basemap(ax,   crs=bocage.crs.to_string(), source=ctx.providers.OpenStreetMap.Mapnik)
ax.set_axis_off()
ax.set_xlim((openfield.x.min(), openfield.x.max()))
ax.set_ylim((openfield.y.min(), openfield.y.max()))
ax.add_artist(ScaleBar(1, location='lower right'))
plt.savefig('openfield_large.png')
```

```python
ax = openfield.plot( color='Black', markersize=1, figsize=(10,10))
ctx.add_basemap(ax, zoom=13,  crs=bocage.crs.to_string(), source=ctx.providers.OpenStreetMap.Mapnik)
ax.set_axis_off()
ax.set_xlim((780000, 790000))
ax.set_ylim((6880000, 6890000))
ax.add_artist(ScaleBar(1, location='lower right'))
plt.savefig('openfield_small.png')
```
